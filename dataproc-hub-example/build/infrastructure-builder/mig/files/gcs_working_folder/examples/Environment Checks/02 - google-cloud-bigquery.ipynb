```python
from google.cloud import bigquery

client = bigquery.Client()
```

#### Public dataset

```python
# Performs a query
QUERY = (
    'SELECT * from `bigquery-public-data.usa_names.usa_1910_2013` '
    'WHERE state = "CA" '
    'LIMIT 5')

# API request
query_job = client.query(QUERY) 
```

```python
# Waits for query to finish
rows = query_job.result()

for row in rows:
    print(row)
```

```python
df = query_job.to_dataframe()
df
```

---
**NOTE**

If you run a queries on private data using `client.query(QUERY)`, your identity should have the following IAM roles or similar:
- **At the project level**: *roles/bigquery.jobUser* (Lower resource is Project) that includes the *bigquery.jobs.create* permission.
- **At the dataset level**: *roles/bigquery.dataViewer* (Lower resource is Dataset) that includes *bigquery.tables.getData* permission.
